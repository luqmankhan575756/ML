```python
import numpy as np
from sklearn.base import BaseEstimator, ClassifierMixin
from sklearn.utils.validation import check_X_y

    
```

```python
class MostFrequentClassClassifier(BaseEstimator, ClassifierMixin):
    # this is constructor in which all the attributes are defined 
    def __init__(self):
        self.most_frequent_ = None   # Attributes ending with an underscore are estimated/learned from data

   
```

```python
#now this learning algorithm will have fit method to learn from data 
def fit(self, x,y):
    
    #validate input x and target vector y
    x, y = check_X_y(x,y)
    
    # validate input X y is 1D 
    y = np.ravel(y)
    # Manually compute the most frequent class
    unique_classes, counts = np.unique(y, return_counts=True)
    self.most_frequent_ = unique_classes[np.argmax(counts)]
    return self

```
