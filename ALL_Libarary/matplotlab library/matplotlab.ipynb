```python
# univariate - Numerical

data = {
    "salary" : [25000,30000,37000,28000.39000,48000,55000,35000,27000,31000]
}
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.shape
```

```python
from matplotlib.lines import lineStyles
# line plot
plt.plot(df["salary"],color = "red",marker = "o", linestyle = "--", linewidth = "2")
plt.grid()
plt.show()
```

```python
# Histograme
plt.hist(df["salary"] , color="green")
plt.show()
```

```python

plt.boxplot(df["salary"])
module: plt
```

```python
df.loc[20] = [0]
```

```python
plt.boxplot(df["salary"])
plt.show()
```

```python
df.drop(index=20, inplace  =True)
```

```python
df.shape
```

```python
df["dept"] = ["HR","IT","CS","CA","HR","IT","CA","HR","IT"]
```

```python
df.head()
```

```python
#pie chart
count = df["dept"].value_counts()
count
```

```python
plt.pie(count, labels=count.index, autopct="%1.1f", explode = [0,0.1,0,0])
plt.axis("equal")
plt.show()
```

```python
#countplot
plt.bar(count.index,count,color = ["green" , "red", "black"])
plt.show()
```
