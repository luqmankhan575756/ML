```python
import pandas as pd 
```

```python
# ✔ Imports the function fetch_openml
# ✔ This function is used to download datasets from OpenML
from sklearn.datasets import fetch_openml
```

```python
df = fetch_openml('titanic', version=1, as_frame=True)['data']
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
import seaborn as sns
```

```python
sns.set()
miss_val_per = pd.DataFrame((df.isnull().sum()/len(df))*100)
```

```python
miss_val_per.plot(kind = 'bar', title = 'missing values in percentage', ylabel = 'percentage')
```

```python
print(f'size of the dataset: {df.shape}')
```

```python
df.drop(['body'], axis = 1 , inplace = True)
print(f"Size of the dataset after dropping a feature: {df.shape}")
```

```python
from sklearn.impute import SimpleImputer
```

```python
print(f'Number of Null values before imputing: {df.age.isnull().sum()}')
```

```python
imp = SimpleImputer(strategy='mean')
df['age'] = imp.fit_transform(df[['age']])
print(f'Number of Null values on age column after imputation: {df.age.isnull().sum()}')
```
